```python
data = pd.read_csv("medicaltranscriptions.csv") 
```

```python
description = data.description
```

```python
description.head()
```

```python
des_str = description.to_string() 
stop_words = set(stopwords.words('english')) 
word_tokens = wordpunct_tokenize(des_str)
word_tokens = [w.lower() for w in word_tokens] 
word_tokens = [w.lower() for w in word_tokens if not w in stop_words] 
word_tokens = [w.lower() for w in word_tokens if w.isalpha()] 
```

```python
word_occurences = pd.Series(word_tokens).value_counts()
word_top10 = word_occurences[:10]
word_top10 # Top 10 Word Occurences in "description"
```

```python
nltk.download('averaged_perceptron_tagger')
pos = nltk.pos_tag(word_tokens)
```

```python
pos
```

```python
pos_noun = []
for i in pos:
    noun = ['NN', 'NNP', 'NNS', 'NNPS']
    if i[1] in noun:
        pos_noun.append(i)
```

```python
pos_n = []
for i in pos_noun:
    pos_n.append(i[0])
pos_n_occurences = pd.Series(pos_n).value_counts()
pos_n_occurences[:10]           #  Top 10 Word Occurences for nouns only
```
